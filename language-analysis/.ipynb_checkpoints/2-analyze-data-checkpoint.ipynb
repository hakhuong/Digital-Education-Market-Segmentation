## Wrangle language lang_df

The Languages column has not been cleaned. However, we only need to find if a language is mentioned in this section. Let's print out a sample of the Language column to see:

In this following section, I will find if country uses these language: 
+ English
+ Spanish
+ French
+ Portugese
+ Chinese 
+ Hindi
+ English used by a majority, with 2 following criteria:
    + English is listed first -- if English is listed first, my assumption is it is popularly used
    + English (official)

```python
# Grab language to match 

# Create columns
lang_df = df.copy()

lang_df['English'] = False
lang_df['Spanish'] = False
lang_df['French'] = False
lang_df['Portuguese'] = False
lang_df['Chinese'] = False
lang_df['Hindi'] = False
lang_df['Arabic'] = False
```

```python
lang_df.sample(5)
```

```python
# Put languages in a list for the below "for" loop 
lang_names = list(lang_df.iloc[:, 2:].columns)
print(lang_names)

# Boolean values for country which has the language name appears in the description
for col in lang_names:
    for row in range(len(lang_df)):
        lang_df.loc[row, col] = (col in lang_df.Languages[row])
        
lang_df.sample(5)
```

```python
# Create column for English majority
lang_df['english_popular'] = False
```

```python
lang_df.loc[1, 'Languages']
```

```python
# Idenfity if English is popular

english_listed_first = '          \n        English'
english_offical = 'English (official'

for row in range(len(lang_df)):
    lang_df.loc[row, 'english_popular'] = (english_listed_first in lang_df.loc[row, 'Languages']) \
                                        or (english_offical in lang_df.loc[row, 'Languages'])
    
lang_df.sample(5)
```

## Examine the frequency of languages used in each countries 

```python
# turn table into the long format 
lang_df_long = lang_df.drop('Languages', axis = 1)
lang_df_long = lang_df_long.melt(id_vars = 'Country', var_name = 'Language')
lang_df_long.sample(5)
```

```python
# Summarize the popularity of languages
lang_df_long[lang_df_long.value == True].groupby('Language').sum().sort_values('value', ascending = False)
```

Turns out that we have many countries are using English as a popular language
