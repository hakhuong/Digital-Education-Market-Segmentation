```python
indicator_list = ['Quality of electricity supply',
                  'Quality of primary education',
       'Primary education enrollment net %',
       'Secondary education enrollment gross %',
       'Tertiary education enrollment gross %',
       'Quality of the education system',
       'Quality of math and science education',
       'Internet access in schools',
        'Individuals using Internet %',
       'Fixed broadband Internet subscriptions', 'Internet bandwidth',
       'Mobile broadband subscriptions',
        'Internet access in schools, 1-7 (best)',
       'Quality of air transport infrastructure, 1-7 (best)',
       'Quality of electricity supply, 1-7 (best)',
       'Quality of management schools, 1-7 (best)',
       'Quality of math and science education, 1-7 (best)',
       'Quality of overall infrastructure, 1-7 (best)',
       'Quality of port infrastructure, 1-7 (best)',
       'Quality of primary education, 1-7 (best)',
       'Quality of railroad infrastructure, 1-7 (best)',
       'Quality of roads, 1-7 (best)',
       'Quality of scientific research institutions, 1-7 (best)',
       'Quality of the education system, 1-7 (best)']
```

```python
df = data[data.Indicator.isin(indicator_list)]
```

```python
df = df[['Country Name', 'Indicator','Subindicator Type', 
    '2007-2008', '2008-2009', '2009-2010', '2010-2011', '2011-2012', '2012-2013', '2013-2014', '2014-2015', 
   '2015-2016', '2016-2017', '2017-2018']]
```

```python
internet_cols = ['Internet access in schools',
        'Individuals using Internet %',
       'Fixed broadband Internet subscriptions', 'Internet bandwidth']
internet_df = df[df.Indicator.isin(internet_cols)]
internet_df.sample(5)
```

```python
internet_df_2 = internet_df[internet_df.Indicator == 'Individuals using Internet %']
internet_df_2 = internet_df_2[internet_df_2['Subindicator Type'] == 'Value']
internet_df_2 = internet_df_2[['Country Name','2017-2018']]
internet_df_2 = internet_df_2.sort_values(by='2017-2018', ascending=False)[26:56]
internet_df_2
```

```python
#graph
sns.set(font_scale = 2)
sns.set_style("ticks")
fig, ax1 = plt.subplots(figsize=(15,6))

#colors
clrs = []
for x in internet_df_2['2017-2018']:
    if x >75:
        clrs.append('green')
    else:
        clrs.append('gray')
        
#chart
ax1 = sns.barplot(x='Country Name', y='2017-2018', data=internet_df_2, palette = clrs)

    
plt.title("Countries ranked 25-55 on % Individuals using Internet", fontsize=24)
ax1.set_ylabel('Percentage', fontsize=16)
ax1.set_xlabel('Country', fontsize=18)

ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45,horizontalalignment='right')
ax1.tick_params(labelsize=15)


plt.show()
```

```python
math_science_quality_df = df[df.Indicator == 'Quality of math and science education']
math_science_quality_df = math_science_quality_df[math_science_quality_df['Subindicator Type'] == '1-7 Best']
math_science_quality_df = math_science_quality_df[['Country Name','2017-2018']]
math_science_quality_df = math_science_quality_df.sort_values(by='2017-2018', ascending=False)
math_science_quality_df = math_science_quality_df[math_science_quality_df['2017-2018'] < 3]
math_science_quality_df
```

```python
#graph
sns.set(font_scale = 2)
sns.set_style("ticks")
fig, ax1 = plt.subplots(figsize=(15,6))


        
#chart
ax1 = sns.barplot(x='Country Name', y='2017-2018', data=math_science_quality_df)

    
plt.title("Countries with Math&Science education quality rated < 3 on 1-7 scale", fontsize=24)
ax1.set_ylabel('Rating (1-7 Best)', fontsize=16)
ax1.set_xlabel('Country', fontsize=18)

ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45,horizontalalignment='right')
ax1.tick_params(labelsize=15)


plt.show()
```
